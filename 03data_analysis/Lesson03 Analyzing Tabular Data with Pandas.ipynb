# Grouping and aggregation

```python
import pandas as pd
```

```python
covid_data = pd.read_csv('italy-covid-daywise.csv')
```

```python
covid_data['date'] = pd.to_datetime(covid_data.date)
```

```python
covid_data['month']  = pd.DatetimeIndex(covid_data.date).month
```

```python
covid_data.groupby('month').sum()  # 使用 sum 聚合 aggregation
```

```python
covid_data.groupby('month')[['new_cases']].sum()
```

```python
covid_data.groupby('month').mean() # 使用 mean() 聚合 
```

```python
covid_data.groupby('month').max() # 使用 max() 聚合 
```

cumsum 累计总和

```python
covid_data['total_cases'] = covid_data.new_cases.cumsum()
```

```python
covid_data['total_deaths'] = covid_data.new_deaths.cumsum()
```

```python
covid_data['total_tests'] = covid_data.new_tests.cumsum()
```

```python
covid_data
```

# Merging data from multiple sources

```python
locations_data = pd.read_csv('locations.csv')
```

```python
locations_data
```

```python
locations_data[locations_data.location == 'Italy']
```

```python
covid_data['location'] = 'Italy'
```

```python
covid_data
```

```python
covid_merged_data = covid_data.merge(locations_data,on='location')
```

```python
covid_merged_data
```

```python
covid_merged_data['cases_per_mill'] = covid_merged_data.total_cases*1e6/covid_merged_data.population               
```

```python
covid_merged_data['deaths_per_mill'] = covid_merged_data.total_deaths*1e6/covid_merged_data.population               
```

```python
covid_merged_data['tests_per_mill'] = covid_merged_data.total_tests*1e6/covid_merged_data.population               
```

```python
covid_merged_data
```
